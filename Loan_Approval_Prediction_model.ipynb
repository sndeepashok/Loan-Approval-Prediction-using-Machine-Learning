```python
#To remove extra spcae before the enire column
```

```python
def clean_data (st):
    st= st. strip()
    return (st)
```

```python
clean_data(' Graduate')
```

```python
## To apply this to enire column
```

```python
data.education = data.education.apply(clean_data)
```

```python
data['education']=data['education'].replace(['Graduate', 'Not Graduate'],[1,0])
```

```python
data.self_employed.unique()
```

```python
data.self_employed = data.self_employed.apply(clean_data)
```

```python
data.self_employed.unique()
```

```python
data['self_employed'] = data['self_employed'].replace(['No', 'Yes'],[0,1])
```

```python
data
```

```python
data.loan_status.unique()
```

```python
data.loan_status = data.loan_status.apply(clean_data)
```

```python
data.loan_status.unique()
```

```python
data['loan_status']= data['loan_status'].replace(['Approved', 'Rejected'],[0,1])
```

```python
data
```

```python
##To select model
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_data=data.drop(columns=['loan_status'])
output_data = data['loan_status']
```

```python
input_data
```

```python
## To split the data using train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_data, output_data, test_size=0.2)
```

```python
output_data
```

```python
x_train.shape,x_test.shape, y_train.shape, y_test.shape
```

```python
## To improve the performance of the data we are scaling the data using standard scaler
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
## To create our model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train_scaled, y_train)
```

```python
model.score(x_test_scaled,y_test)
```

```python
## To Predict some values
```

```python
pred_data= pd.DataFrame([['5','0','1','9800000','24200000','20','382','55000000']],columns=['no_of_dependents','education','self_employed','income_annum','loan_amount','loan_term','cibil_score','Assets'])
```

```python
pred_data= scaler.transform(pred_data)
```

```python
model.predict(pred_data)
```
